```python

priors["add_to_cart_order_mod"] = priors["add_to_cart_order"].copy()
priors["add_to_cart_order_mod"].loc[priors["add_to_cart_order_mod"]>70] = 70
grouped_df = priors.groupby(["add_to_cart_order_mod"])["reordered"].aggregate("mean").reset_index()

plt.figure(figsize=(12,8))
sns.pointplot(grouped_df['add_to_cart_order_mod'].values, grouped_df['reordered'].values, alpha=0.8, color="C0")
plt.ylabel('Reorder ratio', fontsize=12)
plt.xlabel('Add to cart order', fontsize=12)
plt.title("Add to cart order - Reorder ratio", fontsize=15)
plt.xticks(rotation='vertical')
plt.show()
grouped_df.to_csv('reordered_vs_add-to-cart.csv')
del grouped_df
```

# Comparision of number of users with and without days since > 30 order wise

```python

print('loading orders')
orders = pd.read_csv('orders.csv', dtype={
        'order_id': np.int32,
        'user_id': np.int32,
        'eval_set': 'category',
        'order_number': np.int16,
        'order_dow': np.int8,
        'order_hour_of_day': np.int8,
        'days_since_prior_order': np.float32})

grouped = orders.groupby("user_id")['order_number'].aggregate("max").reset_index()

grouped2 = orders.groupby("user_id")['days_since_prior_order'].apply(lambda x : np.count_nonzero(x == 30) ).reset_index()

non_repetetive = grouped.merge(grouped2 , on = 'user_id')

non_repetetive.columns = ["user_id","total_orders","Count_days>30"]

grouped = non_repetetive.groupby("total_orders")['Count_days>30'].aggregate("count").reset_index()

grouped.columns = ['num_of_orders','no_of_users']

grouped2 = non_repetetive.groupby("total_orders")['Count_days>30'].apply(lambda x :np.count_nonzero(x>0)).reset_index()

grouped2.columns = ['num_of_orders','no_of_users_with_days_since>30']

final_df = grouped.merge(grouped2 , on ='num_of_orders')

final_df.to_csv('users_vs_non_repeat_users')
```

```python
final_df.head(5)
```

```python
final_df2 = final_df[final_df.num_of_orders<30]
fig, ax1 = pyplot.subplots(figsize=(10, 10))
tidy = (
    final_df2.set_index('num_of_orders')
      .stack()  # un-pivots the data 
      .reset_index()  # moves all data out of the index
      .rename(columns={'level_1': 'Variable', 0: 'Value'})
)
sns.barplot(x='num_of_orders', y='Value', hue='Variable', data=tidy, ax=ax1)
sns.despine(fig)
```

# Distribution of users based on days since prior order 


```python
group = orders.groupby('user_id')['days_since_prior_order'].aggregate('median').reset_index()


group['binned_days_since'] = pd.cut(group.days_since_prior_order, [0,4,9,12,16,19,24,30],labels = ['<4 days','Weekly (4-9)','9-12 days','Bi-weekly (12-16)','16-19 days','Tri-weekly (19-24)','24-30'])


temp_series = group['binned_days_since'].value_counts()
labels = (np.array(temp_series.index))

df = pd.DataFrame()
df['counts'] = temp_series
df['bins'] = labels
plt.figure(figsize=(20,8))
sns.barplot(labels,temp_series, alpha=0.8 , color = 'C0')
#plt.barplot(sizes, labels=labels, autopct='%1.1f%%', startangle=200)
plt.title("Users Distribution as per reorder habit", fontsize=15)
plt.xticks(rotation='horizontal' , fontsize = 15)
plt.show() 
df.to_csv('binned_users_based_on_days_since.csv')
del temp_series







```
